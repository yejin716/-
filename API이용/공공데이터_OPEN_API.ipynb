```python
import requests 
import json
import pandas as pd 
```

```python
#발급받은 api key 입력하세요.
api_key = 'api key 이용'
```

```python
#사용할 api url 설정
api_url = 'http://apis.data.go.kr/1390802/AgriFood/FdFood/getKoreanFoodFdFoodList'
```

```python
#요청 파라미터 설정

params = {
  'serviceKey' : api_key,
  'service_Type' : 'json', #제공하는 파일형식(xml)
  'Page_No' : '1',
  'Page_Size' : '20',
  'food_Name' : '밥' #검색하고자하는 음식명(옵션)
}
```

```python
#API 요청 보내기 
response = requests.get(api_url, params=params)
```

```python
#응답 데이터 확인 
response
#200 : 정상 수신
```

```python
response.text
```

```python
data = response.json() #json(key,value 형태파일) >> dict로 변경한 데이터 파싱
print(json.dumps(data, indent=4, ensure_ascii=False))
#indent = 4 : 4칸 들여쓰기
#ensure_ascii=False : ASCII문자 비사용(한글이 제대로 출력)
```

```python
#응답 데이터 확인하기 
if response.status_code == 200:
  try:
    data = response.json()
    
    #json 데이터에서 필요한 필드만 추출 >> 리스트 저장 
    items = data['response']['list']
    parsed_data = []
    
    for item in items:
      parsed_data.append({
        'fd_Code' : item['fd_Code'],
        'upper_Fd_Grupp_Nm': item['upper_Fd_Grupp_Nm'],
        'fd_Grupp_Nm': item['fd_Grupp_Nm'],
        'fd_Nm': item['fd_Nm'],
        'fd_Wgh': item['fd_Wgh'],
        'food_Cnt': item['food_Cnt'],
        'food_List': item['food_List']}
      )
      
    #추출한 데이터를 판다스 데이터프레임으로 변환 
    df = pd.DataFrame(parsed_data)
    print(df)
  except json.JSONDecodeError:
    print('Error 발생')
    
```
